```python
# Abstraction :-- It is a process of providing what is required vby the user and hidiing the internal impolementataion
from abc import ABC,abstractmethod
class vehicle(ABC):
    @abstractmethod
    def start(self):
        pass
class car(vehicle):
    def start(self):
        print("Car starts with button")
class bike(vehicle):
    def start(self):
        print("It starts with key")
b1=bike()
b1.start()
b2=car()
b2.start()
```
