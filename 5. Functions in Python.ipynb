```python
#Functions:--It is a block of code that need to be executed continuously in a program
def greet():
    print("Hii")
greet()
```

```python
#Functions with argument
def greet(name):
    print(f"Hii {name}")
greet("Yash")
greet("Sagar")
```

```python
#Functions with keyword arguments
def greet(name,city):
    print(f"Welcome {name} to the {city}")
greet(name='Yash',city='Pune')
```

```python
#Functions with default argument
def greet(name="yash",city="Pune"):
    print("Hii ",name,"Welcomme to ",city)
greet(name="omkar",city="Mumbai")
```

```python
#return function
def greet(first_name,last_name):
    full_name=f"{first_name} {last_name}"
    return full_name
name=greet("Yash","Jagtap")
print(name)
```

```python
#local variables:-- it is a type of variable which is created inside the function and can be accessible wityhin the function

def local():
    info="Be positive"
    print(info)
local()
```

```python
#Global variables:-- it can be accessed within the function and throughout the program

def globe(info):
    print(info)
info="Be positive and happy"
globe("BE")
print(info) #accessed outside the function
```

```python
#Lambda is a anomous and single line functions:--

x=int(input("Enter n1:--"))
y=int(input("Enter n2:--"))

add_no=lambda x,y:x+y
print(add_no(x,y))
```

```python
#Square of no:--

x=int(input("Enter n1:--"))
sq_no=lambda x:x**2
print(sq_no(x))
```
