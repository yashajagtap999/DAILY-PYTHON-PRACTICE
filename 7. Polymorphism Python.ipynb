```python
#Polymorphism with classes
class Bird():
    def sound(self):
        print("Birds make a sound")
class Crow(Bird):
    def sound(self):
        print("crow say caw caw")
class Sparrow(Bird):
    def sound(self):
        print("Sparrow say nothing")
b1=Bird()
b1.sound()
b2=Crow()
b2.sound()
```

```python
class operation():
    def oop(self):
        self.n1=100
        self.n2=100
        print(self.n1+self.n2)
class add(operation):
    def oop(self):
        self.n1=100
        self.n2=100
        print(self.n1-self.n2)
c1=operation()
c1.oop()
c2=add()
c2.oop()        
```
