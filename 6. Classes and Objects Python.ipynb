```python
class car():
    def info(self):
        print("The car is Starting")
    def infoo(self):
        print("The car is stopping")
c1=car()
c1.info()
c1.infoo()
```

```python
# Program with self keyword:-- Self is a special type of keywors used to refer the current object and store the object
class car():
    def info(self,name,brand,color):
        self.name=name
        self.brand=brand # it stores the object
        self.color=color
c1=car()
c1.info("HARRIER","TATA","WHITE")
print(c1.name,c1.brand,c1.color)
```

```python
# taking input from user
class car():
    def inffo(self):
        n1=int(input("enter the number"))
        n2=int(input("Enter teh number"))
        print(n1+n2)
c1=car()
c1.inffo()
c2=car()
c2.inffo()
```

```python
class car():
    def infoo(self,name,brand):
        self.name=name
        self.brand=brand
    def display(self):
        print(f"The car is {self.name}")
        print(f"The brand is {self.brand}")
c1=car()
c1.infoo("XUV700","Mahindra")
c1.display()
c2=car()
c2.infoo("BE6E","Mahindra")
c2.display()
```

```python
''' Constructor :-- It is used to initalize the object automatically without calling it manually
defined by __init__ '''
```

```python
class student():
    def __init__(self,name,grade):
        self.name=name
        self.grade=grade
s1=student("YASH","O")
s2=student("OMKAR","A+")
print(f"The name is {s1.name} and grade is {s1.grade}")
print(f" The name is {s2.name} and grade is {s2.grade}")
```

```python
# Default consructor

class student():
    def __init__(self):
        self.name=input("Enter the name:--")
        self.rolll=int(input("Enter teh roll:--"))
    def infoo(self):
        print("The name is:--",self.name)
        print("The roll is:--",self.rolll)
s1=student()
s1.infoo()
```

```python
#Construcot with default value
class student():
    def __init__(self,name="YASH", roll=11):
        self.name=name
        self.roll=roll
s1=student(name="Omkar",roll=11)
print(s1.name)
print(s1.roll)
```
