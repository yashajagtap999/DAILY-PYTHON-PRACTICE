```python
# Hybrid inheritance
class F():
    def getf(self):
        print("father")
class M():
    def getm(self):
        print("MOther")
class C(M,F):
    def getc(self):
        print("Child")
class GC(C):
    def getgc(self):
        print("GC")
GC=GC()
GC.getgc()
GC.getc()
GC.getf()
GC.getm()
```
