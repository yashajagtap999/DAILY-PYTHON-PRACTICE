```python
#Encapsulation :-- It is a process of binding data and functions toigether to protect the security of data
class bank():
    def __init__(self,acc_no,balance):
        self.acc_no=acc_no
        self.__balance=balance
    def deposit(self,amount):
        self.amount=amount
        self.__balance+=amount
    def display(self):
        print("The balance is:--",self.__balance)
b1=bank(101,1000000000000)
b1.deposit(100)
b1.display()
```
