```python
#Loop :-- for,while
#while :--used to access elements till certain condn is met

i=0
while(i<10):
    print(i)
    i+=1
```

```python
#for :-- Access elements in sequential manner
l1=list(range(0,10,2))
for i in (l1):
    print(l1)
```

```python
#range fun :-- (start,stop,step)
l1=list(range(5,15,2))
l1
```

```python
#nested loop
for i in range(1,4):
    for j in range(2,4):
        print(f'{i},{j}')
```

```python
#break statement :-- It is used to terminate the loop when condn is met
for i in range(1,10):
    if(i==5):
        break
    print(i)
```

```python
#Continue:--it is used to skip current iteration of the loop
for i in range(1,10):
    if(i==5):
        continue
    print(i)
```

```python
# Take input from user
start=int(input("Enter first number"))
stop=int(input("Enter second number"))
skip=int(input("Enter number to skip"))
if(start<stop ):
    print("Valid")
    for i in range(start,stop):
        if i==skip:
            continue
        print(i)
else:
    print("Enter valid number")
```
