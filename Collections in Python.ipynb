```python
# Sequential Collections :-- STring,list,tuple,dictionary,sets
# String
s1="Hello   world"
s2="Hello"+" Guys" #Concatenation
s3="Hii"*4 #Repetation

print(s1.lower())
print(s1.upper())
print(s1.capitalize())
print(s1.title())
print(s1.replace("Hello","Bye"))
print(s1.isalpha())
print(s1.strip())
print(s1[10])
print(s1[0:14:3])
```

```python
# list
l1=[100,102,104,108,109,1001]
l2=[10001,20002]
l4=l1+l2 #concate
l5=[100,200]*4
l1.append(100)
print(l1)
l1.extend([101,103])
print(l1)
l1.sort()
print(l1)
l1.insert(0,11001100)
print(l1)
l1.pop(0)
print(l1)
l1.remove(100)
print(l1)
print(l1.count(100))
print(l1.index(100))
print(min(l1))
print(max(l1))
print(sum(l1))
```

```python
#Tuple :--
T1=(100,100,200,300,200)
T2=(100,900)+(1000,2000)
T3=(100,200)*2
print(T3)
print(T2)
print(T1.index(300))
print(T1.count(100))
```

```python
#Sets:--
s1={800,900,1000,5000}
s2={800,900,100000,500000}
print(s1|s2)
print(s1&s2)
print(s1-s2)
s1.add(100000)
print(s1)
s1.update({800,900})
print(s1)
s1.pop()
print(s1)
s1.remove(900)
print(s1)
s1.clear()
print(s1)
```
