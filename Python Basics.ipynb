```python
# Datatypes :--
l1=100 #int
l2=100.1 #float
l3=10+20j #complex
print(l1+l2+l3)
```

```python
# Opeartors :--
#Arithmetic :-- + ,- ,* ,/ ,// ,**
n1=100
n2=90
print(n1+n2)
print(n1-n2)
print(n1*n2)
print(n1/n2)
print(n1//n2)
print(n1**2)
```

```python
# Comparison operator :-- Always gives output in boolean form
n1=100
n2=90
print(n1>n2)
print(n1==n2)
print(n1<=n2)
print(n1!=n2)
```

```python
# logical operator 
n1=99
n2=88
print(n1>100 and n2>=88)
print(n1>100 or n2>=88)
print(not(n1>n2))
```

```python
# membership operator :-- in,not in
l1=[100,101,102,104,105]
print(100 in l1)
print(109 not in l1)
```

```python
# Identity opeartor :-- is,not is .It is used to compare memory locations
l1=100
l2=l1
l4=200
print(l1 is l2)
print(l1 is l4)
```
