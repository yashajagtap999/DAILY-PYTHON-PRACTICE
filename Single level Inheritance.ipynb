```python
# inheritance :-- it is a process of creating the new object from previous one which has its own characterictics and characteritcs of parent class
```

```python
# single level
class P():
    def base(self):
        print("Its a base class")
class D(P):
    def derived(self):
        print("Its a derived class")
p1=P()
p1.base()
d1=D()
d1.base()
d1.derived()
```
